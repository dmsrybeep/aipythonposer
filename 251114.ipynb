## Pandas 필수 메소드 정리 (Essential Pandas Methods)

Pandas는 파이썬에서 데이터 분석을 위한 강력한 라이브러리입니다. 주로 DataFrame이라는 2차원 테이블 형태의 데이터 구조를 다루며, 데이터 로딩, 탐색, 정제, 조작 및 분석에 필요한 다양한 메소드를 제공합니다.

### 1. 데이터 생성 및 로드 (Data Creation & Loading)

*   `pd.DataFrame()`: 데이터를 사용하여 DataFrame을 생성합니다.
*   `pd.read_csv()`, `pd.read_excel()`, `pd.read_sql()` 등: 다양한 파일 형식에서 데이터를 로드합니다.

### 2. 데이터 탐색 (Data Inspection)

*   `.head(n)`: DataFrame의 상위 `n`개 행을 반환합니다. (기본값 5)
*   `.tail(n)`: DataFrame의 하위 `n`개 행을 반환합니다. (기본값 5)
*   `.info()`: DataFrame의 요약 정보를 제공합니다. (Non-null 값, Dtype 등)
*   `.describe()`: 숫자형 데이터의 기술 통계량(평균, 표준편차, 최소/최대값 등)을 계산합니다.
*   `.shape`: DataFrame의 차원(행, 열)을 튜플로 반환합니다.
*   `.columns`: DataFrame의 열 이름을 반환합니다.
*   `.index`: DataFrame의 인덱스를 반환합니다.
*   `.dtypes`: 각 열의 데이터 타입을 반환합니다.

### 3. 데이터 선택 및 필터링 (Data Selection & Filtering)

*   `df['column']` 또는 `df[['col1', 'col2']]`: 특정 열을 선택합니다.
*   `.loc[row_indexer, column_indexer]`: 라벨 기반으로 데이터를 선택합니다.
*   `.iloc[row_indexer, column_indexer]`: 정수 기반 위치로 데이터를 선택합니다.
*   `df[condition]`: 조건에 따라 행을 필터링합니다.

### 4. 데이터 정제 (Data Cleaning)

*   `.isnull()` / `.isna()`: 각 셀이 결측값(NaN)인지 여부를 나타내는 불리언 DataFrame을 반환합니다.
*   `.notnull()` / `.notna()`: 각 셀이 결측값이 아닌지 여부를 나타내는 불리언 DataFrame을 반환합니다.
*   `.dropna()`: 결측값이 있는 행 또는 열을 제거합니다.
*   `.fillna(value)`: 결측값을 특정 `value`로 채웁니다.
*   `.duplicated()`: 중복된 행을 불리언 Series로 반환합니다.
*   `.drop_duplicates()`: 중복된 행을 제거합니다.

### 5. 데이터 조작 및 변환 (Data Manipulation & Transformation)

*   `.groupby('column')`: 특정 열을 기준으로 데이터를 그룹화합니다.
*   `.agg()`: 그룹화된 데이터에 집계 함수(sum, mean, max 등)를 적용합니다.
*   `.merge()`: 두 DataFrame을 공통 열을 기준으로 병합합니다.
*   `pd.concat()`: 두 개 이상의 DataFrame을 행 또는 열 방향으로 연결합니다.
*   `.apply(func)`: DataFrame의 축(행 또는 열)을 따라 함수를 적용합니다.
*   `.map(dict_or_series)`: Series의 각 값에 매핑 함수를 적용하여 새로운 Series를 반환합니다.
*   `.rename(columns={'old_name': 'new_name'})`: 열 또는 인덱스 이름을 변경합니다.
*   `.sort_values(by='column', ascending=True)`: 특정 열을 기준으로 정렬합니다.
*   `.astype(dtype)`: 데이터 타입을 변경합니다.
*   `pd.to_datetime()`: 문자열을 datetime 객체로 변환합니다.

### 6. 통계 및 요약 (Statistics & Summary)

*   `.mean()`, `.sum()`, `.min()`, `.max()`, `.count()`: 각 열의 통계량을 계산합니다.
*   `.value_counts()`: Series의 고유 값과 그 개수를 반환합니다.
*   `.unique()`: Series의 고유 값을 배열로 반환합니다.
*   `.nunique()`: Series의 고유 값의 개수를 반환합니다.

```python
import pandas as pd
import numpy as np

# 예시 DataFrame 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Alice', 'Frank'],
    'Age': [24, 27, 22, 32, 29, 24, np.nan],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Miami', 'New York', 'Boston'],
    'Salary': [70000, 80000, 60000, 90000, 75000, 70000, 65000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'HR', 'Finance'],
    'DateJoined': ['2020-01-15', '2019-05-20', '2021-03-01', '2018-11-10', '2022-07-01', '2020-01-15', '2023-02-01']
}
df = pd.DataFrame(data)

print("--- Original DataFrame ---")
display(df)
```

### 1. 데이터 탐색 (Data Inspection) 예시

```python
# .head()
print("\n--- df.head() ---")
display(df.head(3))

# .info()
print("\n--- df.info() ---")
df.info()

# .describe()
print("\n--- df.describe() ---")
display(df.describe())

# .shape, .columns, .dtypes
print("\n--- df.shape ---", df.shape)
print("\n--- df.columns ---", df.columns.tolist())
print("\n--- df.dtypes ---")
display(df.dtypes)
```

### 2. 데이터 선택 및 필터링 (Data Selection & Filtering) 예시

```python
# 특정 열 선택
print("\n--- Select 'Name' column ---")
display(df['Name'].head())

# 여러 열 선택
print("\n--- Select 'Name' and 'Salary' columns ---")
display(df[['Name', 'Salary']].head())

# .loc (라벨 기반 선택: 행 인덱스, 열 라벨)
print("\n--- .loc[1:3, ['Name', 'City']] ---")
display(df.loc[1:3, ['Name', 'City']])

# .iloc (정수 위치 기반 선택: 행 인덱스, 열 인덱스)
print("\n--- .iloc[0:4, [0, 2]] ---")
display(df.iloc[0:4, [0, 2]])

# 조건 필터링 (나이가 30세 이상인 사람)
print("\n--- People with Age >= 30 ---")
display(df[df['Age'] >= 30])
```

### 3. 데이터 정제 (Data Cleaning) 예시

```python
# 결측값 확인
print("\n--- .isnull() ---")
display(df.isnull().sum())

# 결측값 제거 (원본 DataFrame 변경 안 함)
df_dropped = df.dropna()
print("\n--- .dropna() (Rows with NaN removed) ---")
display(df_dropped)

# 결측값 채우기 (Age 열의 NaN을 평균으로)
df_filled = df.copy()
df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].mean())
print("\n--- .fillna() (Age NaN filled with mean) ---")
display(df_filled)

# 중복된 행 확인 및 제거
print("\n--- .duplicated() ---")
display(df.duplicated())

df_unique = df.drop_duplicates()
print("\n--- .drop_duplicates() ---")
display(df_unique)
```

### 4. 데이터 조작 및 변환 (Data Manipulation & Transformation) 예시

```python
# 'DateJoined' 열을 datetime 타입으로 변환
df['DateJoined'] = pd.to_datetime(df['DateJoined'])
print("\n--- After converting 'DateJoined' to datetime ---")
display(df.dtypes)

# 새로운 열 생성 (연봉을 월급으로 변환)
df['MonthlySalary'] = df['Salary'] / 12
print("\n--- DataFrame with 'MonthlySalary' ---")
display(df.head())

# .groupby() 및 .agg() (부서별 평균 급여)
print("\n--- Average Salary by Department ---")
display(df.groupby('Department')['Salary'].mean().reset_index())

# .rename() (열 이름 변경)
df_renamed = df.rename(columns={'Name': 'EmployeeName'})
print("\n--- DataFrame with renamed column ---")
display(df_renamed.head())

# .sort_values() (급여 기준으로 내림차순 정렬)
print("\n--- Sorted by Salary (descending) ---")
display(df.sort_values(by='Salary', ascending=False).head())
```

### 5. 통계 및 요약 (Statistics & Summary) 예시

```python
# .value_counts() (도시별 직원 수)
print("\n--- Value Counts for City ---")
display(df['City'].value_counts())

# .unique() (고유한 부서 목록)
print("\n--- Unique Departments ---")
display(df['Department'].unique())

# .nunique() (고유한 부서의 개수)
print("\n--- Number of Unique Departments ---")
display(df['Department'].nunique())
```
